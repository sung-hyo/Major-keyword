# Odds ratio
---
Odds ratio(이하 OR, 교차비, 승산비)는 B가 있을 경우의 A의 확률과 B가 없을 경우의 A의 확률의 비로 정의 되며, A와 B사이의 연관성 정도를 정량화 한 통계이다.

Odds란 Probability(확률)의 다른 표현법으로 볼수 있는데 이진분류(성공/실패, 질환의 발병 등)에 많이 사용된다. 질병의 발병을 예로 들자면, 질병이 발생할 확률이 질병이 발생하지 않을 확률의 몇 배인지를 나타낸다.

$$ \text{odds} = \frac{p(y=1|x)}{1-p(y=1|x)} $$

OR은 한 그룹에서 발생하는 사건의 확률과 다른 그룹에서 발생하는 사건의 확률의 비이다. 각 그룹에서 사건이 발생활 확률을 $p_{1}, \, p_{2}$라고 한다면 다음과 같다. $(1-p_{1} = q_{1},\, 1-p_{2} = q_{2})$

$$ \text{odds ratio} = \frac{p_{1}/(1-p_{1})}{p_{2}/(1-p_{2})} = {p_{1}/q_{1} \over p_{2}/q_{2}}={\frac  {\;p_{1}q_{2}\;}{\;p_{2}q_{1}\;}},$$


### 집단간의 질병 발생
---
어떤 환경에 노출된 집단과 그렇지 않은 집단에서 질병의 발생을 관찰한다고 가정하자. 노출된 집단의 대상자 수 $N_{E}$ 중에 질병이 발생한 대상자 수를 $D_{E}$, 건강한 대상자 수를 $H_{E}$라고 하고, 노출되지 않은 집단의 대상자 수 $N_{N}$ 중에 질병이 발생한 대상자 수를 $D_{N}$, 건강한 대상자 수를 $H_{N}$라고 하자.  아래의 표와 같이 표시하자.

Group|Diseased|	Healthy | Total
---|---|---
Exposed | $$D_{E}$$ | $$H_{E}$$ | $$N_{E}$$
Not exposed | $$D_{N}$$ | $$H_{N}$$ | $$N_{N}$$
Total | $$N_{D}$$ | $$N_{H}$$ | $$N$$

노출 집단에서 질병의 발생률을 $p_{1}$이라고 하면
$$ \text{Exposed odds} = \frac{p_{1}}{1-p_{1}} = \frac{D_{E}/N_{E}}{1-(D_{E}/N_{E})} = \frac{D_{E}/(D_{E}+H_{E})}{1-(D_{E}/(D_{E}+H_{E}))} = \frac{D_{E}}{D_{E}+H_{E}-D_{E}} =\frac{D_{E}}{H_{E}}$$
비노출 집단에서 질병의 발생률을 $p_{2}$이라고 하면
$$ \text{Not exposed odds} = \frac{p_{2}}{1-p_{2}} = \frac{D_{N}/N_{N}}{1-(D_{N}/N_{N})} = \frac{D_{N}}{H_{N}}$$
두 집단에서의 odds ratio는 
$$ {\displaystyle OR={\frac {D_{E}/H_{E}}{D_{N}/H_{N}}}\,, -> OR={\frac {D_{E}H_{N}}{D_{N}H_{E}}}={\frac {D_{E}/D_{N}}{H_{E}/H_{N}}}.}$$
다음과 같이 쓸수 있다.
비노출 집단에 비해 노출집단에서 사건의 일어날 가능성이 OR배 높다라고 해석된다. 

### 예) 흡연 여부와 질병
---

Group|Disease|Normal|total
---|---|---|---
Smoker | 20 | 80 | 100
Non Smoker | 10 | 90 | 100

흡연자에서 질병의 발생 odds는 20/80 = 0.25이고 비흡연자에서 질병발생의 odds는 10/95 = 0.11이다. 
따라서 흡연자와 비흡연자의 odds ratio는 0.25/0.1111 = 2.25이다. 흡연자가 비흡연에 비해 질병에 걸릴 위험이 2.25배 높다라고 해석하다.

## Logistic regression에서의 odds ratio
---
로지스틱 회귀분석은 Cox가 제안한 확률 모델로서 독립변수의 선형 결합을 이용하여 사건의 발생가능성을 예측 하는데 사용되는 통계 기법이다.

![lmvsloglm](https://cdn-images-1.medium.com/max/1600/0*j4b6G61h6FGvaS16.jpg)


로지스틱 함수(logistic function)는 시그모이드 함수(sigmoid function)이며, 입력 t에 대해서 0과 1사이의 값을 출력한다. 로지스틱 함수 $\sigma(t)$는 아래와 같다. 
$$\sigma (t)={\frac {e^{t}}{e^{t}+1}}={\frac {1}{1+e^{-t}}}$$
여기에 입력 $t$를 설명변수가 $x$ 하나인 선형함수라고 놓으면,
$$ t=\beta_{0} + \beta_{1}x$$
로지스틱 함수는 
$${\displaystyle p(x)={\frac {1}{1+e^{-(\beta _{0}+\beta_{1}x)}}}}$$
이며, $p(x)$는 동일한 독립변수 x가 주어졌을 때 실패에 비해 성공(y=1)일 확률로 해석된다. 

로지스틱 함수의 역수($g$의)를 정의 하면
$${\displaystyle g(p(x))=\operatorname {logit} p(x)=\ln \left({\frac {p(x)}{1-p(x)}}\right)=\beta _{0}+\beta _{1}x,}$$
지수화 하면
$$\frac {p(x)}{1-p(x)}=e^{\beta _{0}+\beta _{1}x}$$


독립변수 $X$가 주어졌을대 종속변수 $Y$가 1의 범주에 속할 확률을 의미 한다.$P(Y=1\mid X)$
$$
{\displaystyle {\text{logistic function}} = {\frac {e^{\beta \centerdot X_{i}}}{1+e^{\beta \centerdot X_{i}}}}={\frac {1}{1+e^{-\beta \centerdot X_{i}}}}}
$$

$p(x)$는 예측변수의 선형조합을 고려할때 종속변수가 case(사건)일 확률이다. $p(x)$의 공식은 종속변수가 case(사건)일 확률이 선형회귀식의 로지스틱 함수값과 같다. 


로지스틱 회귀에서는 이진변수에 대해서 odds ratio를 일반화하는 하나의 방법이다. 종속변수 $Y$와 독립변수 $X$가 존재하고, 다른 예측 변수 $Z1, ..., Zp$가 있다고 가정하자. $Y$를 예측하기 위해 다중 로지스틱 회귀분석을 사용하여 $X$에 대한 추정된 회귀계수 $ \hat{\beta}_{x}$는 조건부 odds ratio이다.

$$\exp(\beta _{x})={\frac  {P(Y=1\mid X=1,Z_{1},\ldots ,Z_{p})/P(Y=0\mid X=1,Z_{1},\ldots ,Z_{p})}{P(Y=1\mid X=0,Z_{1},\ldots ,Z_{p})/P(Y=0\mid X=0,Z_{1},\ldots ,Z_{p})}}$$

독립변수가 연속형일경우 1unit당 증가할 때 OR이다. 
$${\displaystyle \mathrm {OR} ={\frac {\operatorname {odds} (x+1)}{\operatorname {odds} (x)}}={\frac {\left({\frac {F(x+1)}{1-F(x+1)}}\right)}{\left({\frac {F(x)}{1-F(x)}}\right)}}={\frac {e^{\beta_{0}+\beta_{1}(x+1)}}{e^{\beta_{0}+\beta_{1}x}}}=e^{\beta_{1}}}$$

${\displaystyle \exp({\hat {\beta }}_{x})}$은 $Z1, ..., Zp$가 고정되었을때 $Y$와 $X$사이의 odds ratio의 추정치로 해석된다.

```python
import statsmodels.api as sm
import numpy as np
X = np.random.normal(0, 1, (100, 3))
y = np.random.choice([0, 1], 100)
result = sm.Logit(y, X).fit()
result.summary2()
```

```python
print(np.exp(result.params))
```

## Incidence Rate Ratio(IRR)
---
역학에서 Rate ratio(발생률, 때로는 incidence density ratio)는 특정 시점에서 발생한 사건의 발생률을 비교하는 데 사용되는 상대적이 차이를 측정방법이다. 

$$ \text{Rate Ratio} = \frac{\text{Incidence Rate 1}}{\text{Incidence Rate 2}} $$

예를 들어 인년(사람의 생) 단위와 같은 person-time에서 사건의 발생이다. 

$$ \text{Incidence Rate} = \frac{\text{events}}{\text{person time}} $$
 
 
### 예
---

Group|Diseased|	Healthy | Person-time(period)
---|---|---
Exposed | $$D_{E}$$ | $$H_{E}$$ | $$T_{E}$$
Not exposed | $$D_{N}$$ | $$H_{N}$$ | $$T_{N}$$

Group|Diseased|	Healthy | Person-time
---|---|---
Exposed | 20 | 80 | 10000
Not exposed | 10 | 90 | 12000

person-time을 100을 놓고 노출그룹에서 incidence rate = 20/100 = 1/5이고 비노출그룹에서 incidence rate = 10/120 = 1/12이다. IRR은 $\frac{1/5}{1/12} = \frac{12}{5}$이 된다. 




## Hazard ratio
---
생존분석에서 hazard ratio(이하 HR)은 두가지 수준의 독립변수(설명변수)의 조건에서 위험률의 비이다. 앞에서 흡연여부 예제에 시간에 따른 HR이라고 한다면 단위시간단 흡연자가 비흡연자에 비해 2.25배의 질병 위험이 있다라고 해석한다. 

### Survival function
---

누적분포함수(cdf) $F(t)$는 일정 시간 t보다 작거나 같은 시간에서 사건을 관측할 확률이며, 이를 일정범위의 시간에 걸쳐 pdf를 통합하면 시간범위에서 관측될 확률을 산출할 수 있다. 
$$F(t)=\int_{0}^{t}f(t)dt$$

$$f(t)={\frac{d}{dt}}F(t)$$


생존함수 $S(t)$는 과거의 t에서 생존할 확률로 기술된다.

$${\displaystyle S(t)=\Pr(T>t)=\int _{t}^{\infty }f(u)\,du=1-F(t).}$$

Kaplan-Meier estimator of survival function
$$\hat{S}(t) = \prod_{t_{(i)} \le t} \frac{n_i - d_i}{n_i}$$

### 예
---

아래의 표는 5명의 대상자에 대해서 관측한 값이다. Censor이 1인 사건의 발생(사망)이다. 

Subject | Time | Censor
---|---|---
1|6|1
2|44|1
3|21|0
4|14|1
5|62|1

생존 일이 가장 짧은 1번 대상의 생존 시간은 6일이며, 이를 이용하여 구간 $I_{0} = \{t: 0 \le t < 6 \} = [0,6\}$ 로 정의 할수 있다. 두번째 생존일은 4번 대상의 14일 이며 구간 $I_{1} = \{t: 6 \le t < 14 \} = [6,14\}$이다. 순서대로 구간을 정의 하고 마지막 구간은 $I_{5} = \{t: t \ge 62 \} = [62, \infty \}$ 이다.

구간 $I_{0}$의 생존률은 1번째 사망자가 6일 이므로 5명 모두 생존해있기 때문에 1.0이다. 6일을 포함한 아주 작은 구간 $(6-\delta, 6]$가 주어졌다면, 이러한 작은 구간에서의 사망의 조건부 확률은 1/5이다. 따라서 생존률은 1-1/5 = 4/5가 된다. 대상자가 6일까지 생존할 확률은 6일 전후의 작은 간격을 통해 생존할 수 있는 조건부 확률로 추정한다. 

$
\hat{S}(6) = 1.0 \times (4/5) = 0.8 \\
\hat{S}(14) = 1.0 \times (4/5) \times (3/4) = 0.6 \\
\hat{S}(21) = 1.0 \times (4/5) \times (3/4) \times (3/3) = 0.6 \\
\hat{S}(44) = 1.0 \times (4/5) \times (3/4) \times (3/3) \times (1/2) = 0.3 \\
\hat{S}(62) = 1.0 \times (4/5) \times (3/4) \times (3/3) \times (1/2) \times (0/1) = 0
$


### Hazard
---
hazard 라는 용어는 시간 t까지 생존한 대상을 조건으로, 시간 t이후의 간격에서 'failure(실패..사망)'의 위험을 기술하기 위한 개념이다.  

$$H(t) = \int_{0}^{t} h(u)du $$

어느 순간의 hazard rate(위험률)는 시간 간격이 0에 가까워짐에 따라 단위 시간 당 사건의 수를 위험 수로 나눈 극한값이다.

$$h(t)=\lim _{{\Delta t\rightarrow 0}}{\frac  {{\mathrm  {observed\;events\;in\;interval}}[t,t+\Delta t]/N(t)}{\Delta t}}$$

Hazard function $h(t)$는 일반적으로 확률밀도함수 f(t)와 생존함수 S(t)를 위험률로 모형화 한것이다. 
$${\displaystyle h(t)=\lim _{dt\rightarrow 0}{\frac {\Pr(t\leq T<t+dt)}{dt\cdot S(t)}}={\frac {f(t)}{S(t)}}}$$

Regression models에서 hazard function은 0보다 큰값이므로 다음과 같은 식으로 hazard function을 매개변수로 지정한다. 
$$h(t)=e^{\beta_{0}}$$
여기에서 독립변수를 추가하면 
$$h(t)=e^{\beta_{0}+\beta_{1}x}$$
이고, 독립변수 x의 log-hazard function은
$$\ln [h(t,x)]=\beta_{0} + \beta_{1}x$$
이다. 

### Hazard ratio
---


$$HR(t, x = a \, versus \, x = b, \beta) = HR(t,a,b,\beta) = \mathrm{exp}[g(t,x=a,\beta)-g(t,x=b,\beta)]=\frac{h(t,a,\beta)}{h(t,b,\beta)}$$

## Cox proportional hazard model
---

Cox의 비레위험모델(proportional hazard model)은 생본모델의 한 종류이다. 비례 위험 가정(proportional hazards condition)은 공변량이 위험과 multiplicatively(곱절, 곱셈) 관련이 있다는 것이다. 

i번째 대상자의 공변량을 $X_i = \{ X_{i1}, \cdots , X_{ip}\}$으로 높으면, Cox의 비례 위험 모형의 위험함수(hazard function)은 다음과 같은 형식을 취한다. 

$${\displaystyle \lambda (t|X_{i})=\lambda _{0}(t)\exp(\beta _{1}X_{i1}+\cdots +\beta _{p}X_{ip})=\lambda _{0}(t)\exp(X_{i}\cdot \beta ).}$$

위의 식은 공변량 벡터(설명변수, 독립변수) $X_{i}$를 갖는 대상 i에게 시간 $t$에서 위험함수를 부여(제공)한다. 

시간 $Y_{i}$에서 대상 i와 함께 관찰 된 사건의 확률은 다음과 같이 쓸 수 있다.

$${\displaystyle L_{i}(\beta )={\frac {\lambda (Y_{i}|X_{i})}{\sum _{j:Y_{j}\geq Y_{i}}\lambda (Y_{i}|X_{j})}}={\frac {\lambda _{0}(Y_{i})\theta _{i}}{\sum _{j:Y_{j}\geq Y_{i}}\lambda _{0}(Y_{i})\theta _{j}}}={\frac {\theta _{i}}{\sum _{j:Y_{j}\geq Y_{i}}\theta _{j}}},}$$

여기서 $\theta_{j} = exp(X_{j} \beta)$이고, 

```python
from lifelines import CoxPHFitter
cph = CoxPHFitter()
cph.fit(dt_dm, 'LENFOL', 'FSTAT')

cph.print_summary()  # access the results using cph.summary
```
